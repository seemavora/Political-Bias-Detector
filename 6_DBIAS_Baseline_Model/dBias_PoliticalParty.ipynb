# Dbias Exploration

```python
import pandas as pd


filtered_df = train_df[train_df['type'].isin(['left', 'center', 'right'])]


type_to_label = {'left': 0, 'center': 1, 'right': 2}
filtered_df['type_label'] = filtered_df['type'].map(type_to_label)


print(filtered_df.head())
```

Step 1: Bias Detection

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments
import torch
from sklearn.model_selection import train_test_split
from datasets import Dataset


df = filtered_df[['text', 'type_label']]

train_texts, val_texts, train_labels, val_labels = train_test_split(
    df['text'].tolist(), df['type_label'].tolist(), test_size=0.2, random_state=42
)


train_dataset = Dataset.from_dict({"text": train_texts, "label": train_labels})
val_dataset = Dataset.from_dict({"text": val_texts, "label": val_labels})


tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")


def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True)

train_dataset = train_dataset.map(tokenize_function, batched=True)
val_dataset = val_dataset.map(tokenize_function, batched=True)


model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased", num_labels=3)


training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    num_train_epochs=3,
    weight_decay=0.01,
)


trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
)


trainer.train()

```

```python
sheet_url = "https://docs.google.com/spreadsheets/d/1PnRunoIW3rXtjKnUA6j-OvqZxsKq4cu9hfTLz0RSOu0/export?format=csv"
try:
    df = pd.read_csv(sheet_url)
    text = df['Text']
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
import re
from transformers import pipeline

fill_mask = pipeline("fill-mask", model="bert-base-uncased")


def recognize_biased_words(text, all_biased_words):
    words = text.split()
    biased_words = [word for word in words if word in all_biased_words]
    return biased_words

def mask_biased_words(text, biased_words):
    for word in biased_words:
        text = re.sub(rf'\b{re.escape(word)}\b', '[MASK]', text)
    return text


def debias_text(masked_text):
    while "[MASK]" in masked_text:

        outputs = fill_mask(masked_text)
        if outputs and isinstance(outputs, list) and "token_str" in outputs[0]:
            top_prediction = outputs[0]["token_str"].strip()

            masked_text = masked_text.replace("[MASK]", top_prediction, 1)
        else:
            masked_text = masked_text.replace("[MASK]", "", 1)
    return masked_text


def process_sentences(sentences, all_biased_words):
    if isinstance(sentences, str):
        sentences = [sentences]

    results = []
    for sentence in sentences:
        biased_words = recognize_biased_words(sentence, all_biased_words)
        masked_text = mask_biased_words(sentence, biased_words)
        debiased_sentence = debias_text(masked_text)
        results.append(debiased_sentence)
    return results if len(results) > 1 else results[0]


sample_sentences = text

processed_sentences = process_sentences(sample_sentences, all_biased_words)
```
